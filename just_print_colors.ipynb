```python
import matplotlib.pyplot as plt
import numpy as np

# Define a list of colors (either names or hexadecimal values)
colors = ['Purple', 'Orange', 'Grey', 'steelblue']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(6, 2))

# Define the distance between patches
distance = 1.0


# Create color patches (using a rectangle for each color)
for i, color in enumerate(colors):
    ax.add_patch(plt.Rectangle((i + i * distance, 0), 1, 1, color=color))

# Set limits and hide axis
ax.set_xlim(0, len(colors) + (len(colors) - 1) * distance)
ax.set_ylim(0, 1)
ax.axis('off')


# Display the color patches
plt.show()
```
