```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a pandas DataFrame
df = pd.read_csv('/home/rittika/Workprojects/iRepertoire/ElementBio/DNA-align-report-analysis_final_results_heavy.csv')

# Set the plot style
sns.set(style="whitegrid")

# Pivot the data to get filenames as rows and metrics as columns
df_pivot = df.set_index('filename')

# Create the heatmap
plt.figure(figsize=(12, 12))
sns.heatmap(df_pivot, annot=True, cmap='YlGn', fmt='g', linewidths=0.5, annot_kws={'size': 12}, cbar_kws={'label': 'Total sequencing reads'})

# Set plot title and labels
plt.title('DNA IRP-ElementBio comparison')
# plt.xlabel('Assay')
plt.ylabel('Dilution')
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
