```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_cell_line_percentages(data_dict, title="Illumina 500ng"):
    """
    Generates a scatter plot comparing cell line percentages for different assays against the Standard assay.

    Parameters:
    - data_dict (dict): Dictionary containing the dataset with assay names as rows and cell line percentages as columns.
    - title (str): Title of the plot.
    """
    df = pd.DataFrame(data_dict)
    df.set_index("Filename", inplace=True)
    
    # Drop 'Standard' to use for X values
    df_non_standard = df.drop("Standard")
    
    # Define color palette for different assays
    assay_colors = plt.cm.tab10.colors  # Unique colors for assays
    markers = "o"  # Use a single marker shape for all
    
    plt.figure(figsize=(8, 6))
    
    # Iterate through each assay and assign a unique color while keeping all cell lines the same color
    for idx, assay in enumerate(df_non_standard.index):
        plt.scatter(
            df.loc["Standard"],  # X values (Standard values)
            df_non_standard.loc[assay],  # Y values (Assay values)
            color=assay_colors[idx % len(assay_colors)],  # Assign unique color per assay
            marker=markers,  # Use the same marker for all assays
            label=assay,
            alpha=0.7
        )
    
    # Plot the Standard row as a line instead of x=y
    plt.plot(df.loc["Standard"], df.loc["Standard"], linestyle="--", color="black", label="Standard", linewidth=1)
    
    plt.xlabel("Standard Percentage per Cell Line")
    plt.ylabel("Cell line percentages")
    plt.xticks(df.loc["Standard"], df.columns, rotation=90)  # Set x-axis labels to cell line names
    plt.title(title)
    
    # Create a custom legend for assays
    legend_patches = [plt.Line2D([0], [0], marker="o", color="w", markerfacecolor=assay_colors[i], markersize=8, label=assay)
                      for i, assay in enumerate(df_non_standard.index)]
    
    plt.legend(handles=legend_patches, title="Assay", bbox_to_anchor=(1.05, 1), loc="upper left")
    
    # Show plot
    plt.show()
```

```python
plot_cell_line_percentages(data_dict=data_500ng)
```
